```python
#Histogram

import numpy as np
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a,color="r",bins=20)
#By default histogram has 10 bins 
plt.title("histogram") 
#plt.show()
#in x axis it has x values 
#y axis it has count or density
```

```python
#Boxplot

data = [np.random.normal(0, std, 10) for std in range(1, 4)]
#(0-mean,std-standard deviation,10-count)
#range(0-1,0-2,0-3)

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False)
#vert=True(default position,vertical or horizontal)
#patch_artist=False(the box disappears empty otherwise default blue color will appear)
data
```

```python
#Pie Chart

# Data to plot

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode= (0, 0, 0.1, 0)  # explode 1st slice.U can also change the other values

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)#auto pct is for int or float,

plt.axis('equal')
plt.show()
```

```python
# Data to plot

labels = 'A', 'B', 'C', 'D'
sizes = [100, 125, 150, 175]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode= (0, 0.2, 0.1, 0)  # explode 1st slice.U can also change the other values

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)#auto pct is for int or float,

plt.axis('equal')
plt.show()
```
