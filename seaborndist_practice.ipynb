```python
# Distribution plots
#distplot,jointplot,pairplot,kdeplot
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
planets=sns.load_dataset("planets")
planets
```

```python
planets.head()
```

```python
planets.tail()
```

```python
#Distribution Plot-Univariate
#distplot

sns.distplot(planets['distance'],color='y')
```

```python
sns.distplot(planets["year"],kde=False,bins=20,color='red')
```

```python
#joint plot-Bivariate

#jointplot() allows you to basically match up two distplots for bivariate data.

â€œscatterâ€
â€œregâ€
â€œresidâ€
â€œkdeâ€
â€œhexâ€
```

```python
sns.jointplot(x="year",y="mass",data=planets,kind="hex",color='red')
```

```python
sns.jointplot(x="year",y="mass",data=planets,kind="reg",color='m')
```

```python
sns.jointplot(x="year",y="mass",data=planets,kind="resid",color='g')
```

```python
sns.jointplot(x="year",y="mass",data=planets,kind="kde",color="k",line_style="dotted")
```

```python
# pairplot
sns.pairplot(planets,hue='method',palette='coolwarm')
```

```python
sns.pairplot(planets,hue='method',palette='pastel')
```

```python
#kdeplot

sns.kdeplot(planets["year"],color="red",linestyle="dashed")
```

```python
sns.kdeplot(planets["year"],color="red",linestyle=":",linewidth=12)
```
