```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y) 
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.plot(x, y)
```

```python
plt.xlabel("speed of the car")
plt.ylabel("age of the car")
plt.title("graph")
plt.plot(y,x,"ko")
```

```python
plt.plot(x,y,"r*",linestyle='dashed',markersize=5)
plt.xlabel('age')
plt.ylabel('speed')
plt.title(' Diagram')
plt.show()
```

```python
plt.plot(x,y,"mh",linestyle='dashed',markersize=5)
plt.xlabel('age')
plt.ylabel('speed')
plt.title(' Diagram')
plt.show()
```

```python
plt.plot(y, marker = 'o', ms = 20)
plt.show()
```

```python
#day two, the age and speed of 15 cars:
x2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x2,y2)
```

```python
plt.plot(x2,y2,'co',linestyle='dashed',ms=5)
plt.xlabel("age")
plt.ylabel("speed")
plt.title("speed by age")
plt.show()
```

```python
plt.scatter(x2,y2,marker='d')
plt.xlabel("age")
plt.ylabel("speed")
plt.title("speed by age")
plt.show()
```

```python
# line plot

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,'r',marker='*',markersize=10)
```

```python
plt.plot(x2,y2,'yo',linestyle='dashed')
plt.plot(y2,x2,'g*',linestyle='dotted')
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("diagram")
plt.show()
```

```python
#BAR PLOT

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y,width =0.2,color='r')
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')
```

```python
#histogram
x = np.random.normal(170, 20, 200)
plt.hist(x,color='y',bins=20)

```

```python
#PIE PLOT


y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.legend(title ="Four Fruits:")
plt.title("pie chart")
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
explode= (0, 0.2, 0.1, 0)
plt.pie(y, labels = mylabels,explode=explode)
autopct='%1f%%'
```
