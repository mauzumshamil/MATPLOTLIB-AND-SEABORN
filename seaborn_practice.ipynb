```python
cars = sns.load_dataset('car_crashes')
cars
```

```python
cars.head()
```

```python
# bar  plot
```

```python
sns.barplot(x='total',y='alcohol',data=cars,color='red')
plt.title("bar graph")
```

```python
sns.barplot(x='not_distracted',y='no_previous',data=cars,palette='coolwarm')
```

```python
import numpy as np
sns.barplot(x='total',y='speeding',data=cars,palette='Set1',estimator=np.std)
```

```python
# count plot
```

```python
sns.countplot(x='abbrev',data=cars)
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(x='sex',data=titanic,hue='sex',palette='Set1')
```

```python
sns.barplot(x='sex',y='age',palette='coolwarm',hue="sex",data=titanic,estimator=np.std)
```

```python
#box plot
```

```python
sns.boxplot(x='sex',y='fare',data=titanic,palette='coolwarm')
```

```python
sns.boxplot(x='sex',y='pclass',data=titanic,orient='h',palette='Set1')
```

```python
titanic.tail()
```

```python
sns.barplot(x='fare',y='age',data=titanic,orient='h',hue='alone',palette='coolwarm')
```

```python
#violinplot
```

```python
sns.violinplot(x='fare',y='age',data=titanic,hue='alone',palette='coolwarm')
```

```python
sns.violinplot(x='class',y='fare',palette='Set2',data=titanic)
```

```python
sns.violinplot(x='class',y='fare',palette='Set2',data=titanic,hue='alone')
```

```python
sns.violinplot(x='class',y='fare',palette='Set2',data=titanic,hue='alone',split='true')
```

```python
#stripplot
```

```python
sns.stripplot(x='class',y='fare',palette='Set2',data=titanic)
```

```python
sns.stripplot(x='class',y='fare',palette='Set2',data=titanic,jitter=True,hue="sex")
```

```python
#swarmplot
```

```python
sns.swarmplot(x='class',y='fare',palette='Set2',data=titanic,hue='alone')
```

```python
#violin x swarm
```

```python
sns.violinplot(x='class',y='fare',palette='Set2',data=titanic)
sns.swarmplot(x='class',y='fare',palette='Set2',data=titanic,hue='alone')
```
