```python
#import library
import seaborn as sns
#%matplotlib inline
```

```python
import seaborn as sns
print(sns.get_dataset_names())
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
tips.head()
```

```python
#Distribution Plot-Univariate
sns.distplot(tips['total_bill'])
```

```python
#usually dist plots can be default view by histogram and kde plt
#histogram refers-the frequency levels of the observations
#kde refers-the probability chances occurs of the observations
```

```python
sns.distplot(tips['tip'])
```

```python
#To remove the kde layer and just have the histogram use:kde=False
sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
#joint plot-Bivariate

#jointplot() allows you to basically match up two distplots for bivariate data.

“scatter”
“reg”
“resid”
“kde”
“hex”
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#joint plots are combination of histogram(2) and scatter plots
#scatter plot refers-combination of total bill and tip
#histogram refers-total bill in x-axis and tip in y-axis
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#by using "kind=hex" maximum frequency denotes dark shades
#also easy to understand abt the frequencies obtained
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#Pair plot-pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).

sns.pairplot(tips)
```

```python
pair plot contains combination of histogram,scatter and strip plots
histogram is univariate so histogram refers total bill,size,tip in x-axis for 3 histograms
scatter and strip plots are bivariate,it refers x-axis and y-axis for their respectives
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')

#hue represents bivariate observations
```

```python
#kdeplot-
#kdeplots are Kernel Density Estimation plots. These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value

sns.kdeplot(tips["total_bill"])
```
